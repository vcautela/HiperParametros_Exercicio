```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
df = pd.read_excel('/content/house.xlsx')

df.head()
```

```python
df = pd.get_dummies(df)
df.head()
```

```python
df.info()
```

```python
df = df.dropna()
df.info()
```

```python
target = np.array(df['SalePrice'])
features = df.drop("SalePrice", axis=1)

features_list = list(features.columns)

features = np.array(features)
```

```python
train_features, test_features, train_target, test_target = train_test_split(features, target, test_size=0.25, random_state=42)
```

## Treinando com o GradientBoostingRegressor

```python
gbr = GradientBoostingRegressor(n_estimators=100)
gbr.fit(train_features, train_target)
gbr_pred = gbr.predict(test_features)
```

```python
r_sq = gbr.score(test_features, test_target)
print(r_sq)

print("MAE:", metrics.mean_absolute_error(test_target, gbr_pred))
print("MSE:", metrics.mean_squared_error(test_target,gbr_pred))
print("RMSE:", np.sqrt(metrics.mean_squared_error(test_target,gbr_pred)))
```

## GridSearch


```python
parameters = {"learning_rate" : [0.03],
              "subsample" : [0.2],
              "n_estimators" : [100, 1000, 1500],
              "max_depth" : [5, 8]
}

grid_search = GridSearchCV(gbr, parameters, scoring = 'r2', cv = 2, n_jobs=-1)
```

```python
grid_search.fit(train_features, train_target)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
```
